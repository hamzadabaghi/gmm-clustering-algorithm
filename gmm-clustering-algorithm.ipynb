```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pandas import DataFrame
from sklearn import datasets
from sklearn.mixture import GaussianMixture
```

Loading datasets

```python
iris = datasets.load_iris()
```

Visualizing data

```python
X = iris.data[:, :2]
d = pd.DataFrame(X)
plt.scatter(d[0],d[1])
plt.show()
```

```python
gmm = GaussianMixture(n_components = 3)
# Fit the GMM model for the dataset
# which expresses the dataset as a
# mixture of 3 Gaussian Distribution
gmm.fit(d)
# Assign a label to each sample
labels = gmm.predict(d)
d['labels']= labels
d0 = d[d['labels']== 0]
d1 = d[d['labels']== 1]
d2 = d[d['labels']== 2]
# plot three clusters in same plot
plt.scatter(d0[0], d0[1], c ='r')
plt.scatter(d1[0], d1[1], c ='yellow')
plt.scatter(d2[0], d2[1], c ='g')
plt.show()
```

```python
from sklearn.mixture import GaussianMixture
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
import glob
```

Preprocessing

```python
def Preprocessing(path):
    list_image = []
    path = glob.glob(path)
    for path_img in path:
        # reading the image
        img=plt.imread(path_img)
        # resizing
        img = cv.resize(img,(60,40))
        # binarization
        (thresh, im_bw) = cv.threshold(img, 128, 1, cv.THRESH_BINARY )
        img=np.array(im_bw)
        # Flattening the image
        img=img.flatten()
        list_image.append(img)
    return list_image
```

Creating a Gaussien Mixture model

```python
# paths
path_TEST = "./DataToPredict/*.jpg"
path_train = "./DataToLearn/*.jpg"
# the gmm model
model_gmm = GaussianMixture(n_components =2,covariance_type='diag',init_params='kmeans')
```

Fitting

```python
def gmm_learn():
    images = Preprocessing(path_train)
    model_gmm.fit(images)
```

Clustering

```python
def classification(path):
    classes = []
    k =1
    path = glob.glob(path)
    plt.figure(figsize=(60,40))
    for path_img in path:
        # reading the image
        img=plt.imread(path_img)
        # resizing
        img = cv.resize(img,(60,40))
        # binarization
        (thresh, im_bw) = cv.threshold(img, 128, 1, cv.THRESH_BINARY )
        img=np.array(im_bw)
        # Flattening the image
        image=img.flatten().reshape(1,-1)
        if(k<=10):
            img = plt.imread(path_img)
            plt.subplot(2,5,k)
            plt.imshow(img,cmap='gray')
            class_ = model_gmm.predict(image)
            plt.title("class"+str(class_),fontsize=60)
            classes.append(class_)
        k=k+1
```

```python
gmm_learn()
data = classification(path_TEST)
plt.show()
```
